```python
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

from patient_simulator.misc.plotting import save_to_figures

plt.style.use(Path("../patient_simulator/misc/style.mplstyle"))
```

```python
categories = [
    "Full\ninformation",
    "Current\nsimulators",
    "Real\nHumans",
    "PWP\n(Ours)",
]
values = [82, 79, 48, 46]
fills = ["#808080", "#DD8452", "#555555", sns.color_palette("Dark2")[0]]

fig, ax = plt.subplots(figsize=(6, 2))

x = list(range(len(categories)))
ax.bar(x, values, color=fills, edgecolor="none", width=0.62, zorder=2)

ax.set_ylim(0, 80)
ax.set_xticks(x)
# ax.set_ylabel("Benchmarked LLM\nPerformance")
ax.set_xticklabels([])
ax.set_yticks([0, 25, 50, 75])
ax.tick_params(axis="x", length=0)
ax.tick_params(axis="y", length=0, labelleft=False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_color("0.6")
ax.grid(axis="y", linestyle="--", linewidth=0.5, color="0.7", zorder=0)
ax.set_axisbelow(True)

fig.tight_layout()
save_to_figures(fig, "motivation_disposition_accuracy", dpi=300)
plt.show()
```
